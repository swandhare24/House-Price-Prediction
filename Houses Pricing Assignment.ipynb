### Start feature selection pipeline
Categorical Feature : Ordinal Encoder + StandardScaler

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1 = Pipeline(steps=[('imputer',SimpleImputer(strategy='mean')),
                            ('scaler',StandardScaler())])

cat_pipe1 = Pipeline(steps=[('imputer',SimpleImputer(strategy='constant', fill_value='NotAvailable')),
                            ('ordinal',OrdinalEncoder()),
                            ('scaler',StandardScaler())])

pre1 = ColumnTransformer([('num',num_pipe1, con),
                          ('cat',cat_pipe1, cat)])
```

```python
X_pre = pre1.fit_transform(X)
X_pre
```

```python
cols = pre1.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre, columns=cols)
X_pre.head()
```

### Feature Selection

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
sel = SequentialFeatureSelector(lr, direction='backward', n_jobs=-1)
sel_features = sel.fit_transform(X_pre, Y)
```

```python
imp_cols = sel.get_feature_names_out()
imp_cols
```

```python
sel_cols = []
for i in imp_cols:
    s = i.split('__')[1]
    sel_cols.append(s)
```

```python
sel_cols
```

```python
X_sel = X[sel_cols]
X_sel.head()
```

### Cat, Con sep

```python
cat_sel, con_sel = catconsep(X_sel)
```

```python
cat_sel
```

```python
con_sel
```

### Create a final pipeline
Categorical - OneHotEncoded

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = Pipeline(steps=[('imputer',SimpleImputer(strategy='mean')),
                            ('scaler',StandardScaler())])

cat_pipe2 = Pipeline(steps=[('imputer',SimpleImputer(strategy='constant', fill_value='NotAvailable')),
                            ('OHE',OneHotEncoder(handle_unknown='ignore'))])

pre2 = ColumnTransformer([('num',num_pipe2,con_sel),
                          ('cat',cat_pipe2,cat_sel)])
```

```python
X_sel_pre = pre2.fit_transform(X_sel).toarray()
X_sel_pre
```

```python
final_cols = pre2.get_feature_names_out()
final_cols
```

```python
X_sel_pre = pd.DataFrame(X_sel_pre,columns=final_cols)
X_sel_pre.head()
```

### Train Test split
Train : 80% Test : 20%

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_sel_pre, Y, test_size=0.2, random_state=21)
```

### Create a base model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain, ytrain)
```

```python
lr.score(xtrain, ytrain)
```

```python
lr.score(xtest, ytest)
```

```python
from PM6 import evaluate_model
evaluate_model(xtrain, ytrain, xtest, ytest, lr)
```

### Final model building

### Ridge model

```python
import numpy as np
alphas = np.arange(0.1, 100, 0.1)
alphas
```

```python
params = {'alpha':alphas}
params
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()
gscv = GridSearchCV(ridge, param_grid=params, cv=5, scoring='neg_mean_squared_error')
gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```
